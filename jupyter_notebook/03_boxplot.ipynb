```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
kashti = seaborn.load_dataset("titanic")
sns.boxplot(x = "sex",
               y="fare", showmeans=True,
            meanprops = {"marker":"+",
                        "markersize":"12",
                        "markeredgecolor":"red"},
               data=kashti)
plt.xlabel("how many survived")
plt.title("kitny bach gaye")
plt.show()
```
