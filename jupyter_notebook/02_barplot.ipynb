```python
# only change the x and y axis values, in placce of ci can also be used the word error bars
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean 
# Load data
kashti = sns.load_dataset("titanic")
kashti
# draw a barplot
sns.barplot(x = "fare", y = "class", hue = "who", data = kashti,saturation=1, ci= None)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean 
# Load data
kashti = sns.load_dataset("titanic")
# draw a barplot
sns.barplot(x = "class", y = "fare", data = kashti,
           linewidth=3, facecolor=(1,1,1,0.2),
           errcolor="0.5",edgecolor="0")
plt.show()
```
