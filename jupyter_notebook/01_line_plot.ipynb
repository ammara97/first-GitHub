# Import libraries
seaborn automatically install these libraries

- numpy
- scipy
- pandas
- matplotlib

```python
# Import libraries 
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)
plt.show()
```

# Adding titles

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)
plt.title("phoolon ka guldasta")
plt.show()
```

## adding the limit

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)
plt.title("phoolon ka guldasta")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## changing the size of figure

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load data
phool = sns.load_dataset("iris")
# size of figure
plt.figure(figsize=(10,8))
# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)
plt.title("phoolon ka guldasta")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load data
phool = sns.load_dataset("iris")
# size of figure, ham kisi b size ki figure bna sakty
plt.figure(figsize=(4,2))
# draw a line plot
sns.lineplot(x = "sepal_length", y = "sepal_width", data = phool)
plt.title("phoolon ka guldasta")
plt.xlim(2)
plt.ylim(1)
plt.show()
```
